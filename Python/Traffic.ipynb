```python
from sklearn.preprocessing import MinMaxScaler
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import cut_tree
from sklearn.cluster import KMeans
```

```python
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(df)
```

```python
# Elbow curve method to find the ideal number of clusters.


scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(df)

clusters=list(range(2,8))
ssd = []
for num_clusters in clusters:
    model_clus = KMeans(n_clusters = num_clusters, max_iter=150,random_state= 50)
    model_clus.fit(data_scaled)
    ssd.append(model_clus.inertia_)

plt.plot(clusters, ssd);
```

```python
from sklearn.metrics import silhouette_score
range_n_clusters = [2, 3, 4, 5, 6, 7, 8]

for num_clusters in range_n_clusters:
    
    # intialise kmeans
    kmeans = KMeans(n_clusters=num_clusters, max_iter=50,random_state= 100)
    kmeans.fit(data_scaled)
    
    cluster_labels = kmeans.labels_
    
    # silhouette score
    silhouette_avg = silhouette_score(data_scaled, cluster_labels)
    print("For n_clusters={0}, the silhouette score is {1}".format(num_clusters, silhouette_avg))
```

```python
# 假設 df 是包含特徵的資料框
# 計算特徵之間的相關係數矩陣
corr_matrix = df.corr()

# 使用 Seaborn 的 heatmap 函數繪製相關係數熱圖
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')

# 找到相關性最高的兩個特徵
# 使用 unstack 方法將相關係數矩陣轉換為序列
corr_series = corr_matrix.unstack()

# 排除對角線上的相關係數（自身與自身的相關性為 1）
corr_series = corr_series[corr_series.index.get_level_values(0) != corr_series.index.get_level_values(1)]

# 找到相關性最高的兩個特徵
max_corr = corr_series.abs().max()
feature1, feature2 = corr_series[corr_series.abs() == max_corr].index.tolist()[0]

# 輸出相關性最高的兩個特徵
print(f"最高相關性特徵1: {feature1}")
print(f"最高相關性特徵2: {feature2}")
```

```python
import pandas as pd

# 假設 df 是包含特徵的資料框
# 計算特徵之間的相關係數矩陣
corr_matrix = df.corr()

# 輸出特徵值相關性數值表格
corr_table = pd.DataFrame(corr_matrix.stack(), columns=['相關性'])
corr_table.index.names = ['特徵1', '特徵2']
corr_table.reset_index(inplace=True)

# 找到相關性最高的兩個特徵
max_corr = corr_table['相關性'].abs().max()
max_corr_row = corr_table[corr_table['相關性'].abs() == max_corr].iloc[0]

feature1 = max_corr_row['特徵1']
feature2 = max_corr_row['特徵2']
correlation = max_corr_row['相關性']

# 輸出相關性最高的兩個特徵及其相關性數值
print(f"最高相關性特徵1: {feature1}")
print(f"最高相關性特徵2: {feature2}")
print(f"相關性數值: {correlation}")

# 輸出特徵值相關性數值表格
print(corr_table.to_string(index=False))
```

```python
import pandas as pd

# 假設 df 是包含特徵的資料框
# 計算特徵之間的相關係數矩陣
corr_matrix = df.corr().abs().round(4)

# 轉換相關係數矩陣為DataFrame
corr_df = pd.DataFrame(corr_matrix.stack(), columns=['相關性'])

# 重設索引，使特徵成為列
corr_df.reset_index(level=[0, 1], inplace=True)

# 重新命名列名稱
corr_df.columns = ['特徵1', '特徵2', '相關性']

# 選取特徵一為'ACCIDENT_DEAD'和'ACCIDENT_INJURY'的結果
corr_df_filtered = corr_df[(corr_df['特徵1'] == 'ACCIDENT_DEAD') | (corr_df['特徵1'] == 'ACCIDENT_INJURY')]

# 根據相關性絕對值進行排序
corr_df_filtered = corr_df_filtered.sort_values(by='相關性', ascending=False)

# 輸出結果
print(corr_df_filtered.to_string(index=False, float_format='{:.4f}'.format))

# 輸出結果到CSV檔案
corr_df_filtered.to_csv('correlation_matrix_filtered.csv', index=False)
```
